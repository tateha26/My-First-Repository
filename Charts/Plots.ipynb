```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
clean_merge_csv = "../Resources/merged_data/clean_merge2.csv"
```

```python
clean_merge_df = pd.read_csv(clean_merge_csv)
clean_merge_df.head()
```

```python
cm_group_pay = clean_merge_df.groupby(["Year", "Industry"]).mean()[["Annual Average Employment","Annual Average Pay"]]
cm_group_pay.round()
```

```python
cm_group_price = clean_merge_df.groupby(["Year"]).mean()["Avg_Price"]
cm_price_df = pd.DataFrame(cm_group_price)
cm_price_df
```

```python
fig, ax=plt.subplots()
ax.plot(cm_price_df, marker = 'o', label = "Average Price")
    


ax.set_title("National Average Change in Housing Prices")
ax.set_xlabel("Time (years)")
ax.set_ylabel("Cost (USD)")
ax.set_ylim(100000, 180000, 50000)

ax.yaxis.grid(True, ls='--', alpha=.5)
```

```python
# Save figure
plt.savefig("National_Avg_Housing_Prices.png")
plt.show()
```
