```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Pull in file 
merged_df = pd.read_csv('Resources/clean_merge.csv')

merged_df
```

```python
merged_df['Industry'] = merged_df['Industry'].str.replace('\d+', '')
merged_df['Industry'] = merged_df['Industry'].str.strip()

merged_df
```

```python
industry_group = merged_df.groupby(['Year', 'State', 'Industry'])
avg_employ = industry_group["Annual Average Employment"].mean()
avg_employ = pd.DataFrame(avg_employ).reset_index()
avg_employ.head()
```

```python
avg_year = avg_employ.groupby(['Year','Industry'])
avg_EmployYear = avg_year['Annual Average Employment'].mean()

avg_EmployYear = pd.DataFrame(avg_EmployYear).reset_index()
avg_EmployYear
```

```python
avg_EmployYear.min()
```

```python
avg_EmployYear.max()
```

```python
construction_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Construction", "Annual Average Employment"]
education_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Education and health services", "Annual Average Employment"]
finance_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Financial activities", "Annual Average Employment"]
goods_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Goods-producing", "Annual Average Employment"]
information_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Information", "Annual Average Employment"]
leisure_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Leisure and hospitality", "Annual Average Employment"]
manufacturing_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Manufacturing", "Annual Average Employment"]
natural_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Natural resources and mining", "Annual Average Employment"]
other_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Other services", "Annual Average Employment"]
professional_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Professional and business services", "Annual Average Employment"]
service_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Service-providing", "Annual Average Employment"]
total_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Total, all industries", "Annual Average Employment"]
trade_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Trade, transportation, and utilities", "Annual Average Employment"]
unclassified_avg = avg_EmployYear.loc[avg_EmployYear['Industry'] == "Unclassified", "Annual Average Employment"]

years = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
```

```python
construction = plt.plot(years, construction_avg, color="red", label="Construction")
education = plt.plot(years, education_avg, color="maroon", label="Education and health services")
finance = plt.plot(years, finance_avg, color="yellow", label="Financial activities")
goods = plt.plot(years, goods_avg, color="lime", label="Goods-producing")
information = plt.plot(years, information_avg, color="green", label="Information")
leisure = plt.plot(years, leisure_avg, color="magenta", label="Leisure and hospitality")
manufacturing = plt.plot(years, manufacturing_avg, color="purple", label="Manufacturing")
natural = plt.plot(years, natural_avg, color="cyan", label="Natural resources and mining")
other = plt.plot(years, other_avg, color="teal", label="Other services")
professional = plt.plot(years, professional_avg, color="blue", label="Professional and business services")
service = plt.plot(years, service_avg, color="navy", label="Service-providing")
total = plt.plot(years, total_avg, color="silver", label="Total, all industries")
trade = plt.plot(years, trade_avg, color="gray", label="Trade, transportation, and utilites")
unclassified = plt.plot(years, unclassified_avg, color="black", label="Unclassified")

plt.xlabel("Year")
plt.ylabel("Annual Average Employment")

tick_locations = years
plt.xticks(np.arange(2009, 2019, 1))
plt.yticks(np.arange(0, 70000, 5000))

#plt.legend(['finance', 'professional'], loc="best")
#plt.legend(['construction', 'education', 'finance', 'goods', 'information', 'leisure', 'manufacturing', 'natural', 'other', 'professional', 'service', 'total', 'trade', 'unclassified'], loc="best")
#plt.legend('Construction', 'Education and health services', 'Financial activities', 'Goods-proudcing', 'Information', 'Leisure and hospitality', 'Manufacturing', 'Natural resources and mining', 'Other services', 'Professional and business services', 'Service-providing', 'Total', 'Trade, transportation, and utilites', 'Unclassified')
plt.show()
```
