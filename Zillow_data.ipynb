```python
#Import needed libraries
import pandas as pd
```

```python
#Define locations of datafiles using glob 
zillow_data = ('resources/Zillow_Data/County_Zhvi_AllHomes.csv')
```

```python
df_zillow = pd.read_csv(zillow_data, encoding="ISO-8859-1")
```

```python
df_zillow.head()
```

```python
#Remove years not between 2009 and 2018

yr = '1996'
while int(yr) < 2009:
    df_zillow = df_zillow.drop(df_zillow.filter(like=yr), axis=1)
    yr=str(int(yr)+1)
    
df_zillow = df_zillow.drop(df_zillow.filter(like='2019'), axis=1)
df_zillow['FIPS'] = df_zillow['FIPS'].apply(lambda x: '{0:0>5}'.format(x))
df_zillow
```

```python
yr = '2009'
while int(yr) < 2019:
    yr_col = [col for col in df_zillow if col.startswith(yr)]
    df_zillow["Avg_Price"+yr] = ""
    for index, row in df_zillow.iterrows():
        avg = row.loc[yr_col].mean(axis=0)
        df_zillow.loc[index, "Avg_Price"+yr] = avg
    yr=str(int(yr)+1)
# df_zillow.head()
```

```python
# Remove extra month columns

yr = '2009'
while int(yr) < 2019:
    df_zillow = df_zillow.drop(df_zillow.filter(like=yr+"-"), axis=1)
    yr=str(int(yr)+1)
    
df_zillow
```

```python
df_zillow1 = pd.wide_to_long(df_zillow, stubnames='Avg_Price', i=['RegionName','State','FIPS'], j='Year')
df_zillow1 = df_zillow1.drop(['SizeRank','RegionID','StateCodeFIPS','MunicipalCodeFIPS','Metro'], axis=1)
df_zillow1
```

```python
df_zillow1.reset_index()
```

```python
df_zillow1.to_csv('resources/Zillow_Data/Avg_Home_County.csv')
```
