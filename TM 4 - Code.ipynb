```python
#Import needed libraries
import pandas as pd
import glob 
```

```python
#Define locations of datafiles using glob 
employment_file_to_load = glob.glob('Resources/blsdata/*.xlsx')
bls_fips_file = ('Resources/blsdata/area_titles.csv')
print(employment_file_to_load)
print(bls_fips_file)
```

```python
#Read in BLS Data by passing in list of files in resources/blsdata and concatanating to dataframe


#test_df = pd.read_excel('Resources/blsdata\\allhlcn09.xlsx')
#test_df.head()
df_from_each_bls_file = (pd.read_excel(f) for f in employment_file_to_load)
concatenated_bls_df   = pd.concat(df_from_each_bls_file, ignore_index=True)




concatenated_bls_df.head()


    
    
```

```python
#write employment data to csv

concatenated_bls_df.to_csv('Resources/blsdata/bls_concat.csv')
```
